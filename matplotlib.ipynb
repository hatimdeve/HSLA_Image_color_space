```python
plt.style.use('seaborn')
fig=plt.figure()
ax=plt.axes()
x=np.linspace(1,10,1000)
y=np.sin(x)
y1=np.cos(x)
y2=y+(y1-0.5)
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.style.use('seaborn')
fig=plt.figure(figsize=(4,3))
ax=plt.axes()
x=np.linspace(1,10,1000)
y=np.sin(x)
y1=np.cos(x)
y2=y+(y1-0.5)
plt.plot(x,y,color='blue',linestyle='solid')
plt.plot(x,y1,color='g',linestyle='dashed')
plt.plot(x,y2,color=(1,0,0),linestyle='dotted')
```

```python
from sklearn.datasets import fetch_lfw_people
faces =fetch_lfw_people(min_faces_per_person=70)
```

```python
fig,ax=plt.subplots(4,5,figsize=(4,3),sharex='col',sharey='row')
k=0
for i in range(4):
  for j in range(5):
    ax[i,j].imshow(faces.images[k],plt.cm.bone)
    ax[i,j].set(xticks=[],yticks=[])
    k+=1
```

```python
from mpl_toolkits.mplot3d import  Axes3D
fig=plt.figure(figsize=(4,3))
ax=fig.gca(projection='3d')
theta=np.linspace(-4*np.pi,4*np.pi,100)
z=np.linspace(-2,2,100)
r=z**2+1
x=r*np.sin(theta)
y=r*np.cos(theta)
ax.plot(x,y,z,label='3d curve')
ax.legend()
```

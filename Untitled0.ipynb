```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#des donnes sous forme de liste
prix_ms=[30,56,88,97,103,110]
prix_mac=[80,95,100,120,130,150]
annee=[2015,2016,2017,2018,2019,2020]
```

```python
plt.scatter(annee,prix_ms,label="prix windows",c="r")
plt.plot(annee,prix_mac,label="prix mac",c="b")
plt.legend()
plt.xlabel("Annee")
plt.ylabel("prix")
plt.title(" evolution prix pour les 5 derniere anne")
```

```python
plt.subplot(1,2,1)
plt.scatter(annee,prix_ms,label="prix windows",c="r")
plt.subplot(1,2,2)
plt.plot(annee,prix_mac,label="prix mac",c="b")
```

```python
index_annee=np.arange(len(annee))
a=0.25
plt.bar(index_annee,prix_ms,label="prix windows",width=0.25)
plt.bar(index_annee+a,prix_mac,label="prix mac",width=0.25)
plt.legend()
plt.xlabel("Annee")
plt.ylabel("prix")
plt.title(" evolution prix pour les 5 derniere anne")
```
